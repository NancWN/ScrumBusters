```python
import requests
import json
import pandas as pd
import numpy as np
```

```python

sample = "pump_evx_20211024_binance__simplified"      # enter pairs you are looking for
path = "C:\\Users\\Ahmet\\Desktop\\Uni\\TechLab\\Scam Busters\\Samples\\financial_data\\" # the path to your folder with the data

df = pd.read_csv(path + sample + '.csv')    # reads the csv

# turnes unix time to datetime
df['open_time'] = pd.to_datetime(df['open_time'],unit='ms')

df
```

```python
# feature 1: trade volume change between 2 days in percent

#day 1
start_date = pd.to_datetime('2021-10-20 18:00:00')
end_date = pd.to_datetime('2021-10-20 19:00:00')
#day 2
start_date2 = pd.to_datetime('2021-10-21 18:00:00')
end_date2 = pd.to_datetime('2021-10-21 19:00:00')

#data

d01 = df.loc[(df['open_time'] > start_date) & (df['open_time'] < end_date)]
d02 = df.loc[(df['open_time'] > start_date2) & (df['open_time'] < end_date2)]

f01 = np.sum(d01['volume'])/np.sum(d02['volume'])*100

print('volumen Tag 1 = ', np.sum(d01['volume']))
print('volumen Tag 2 = ', np.sum(d02['volume']))
print('feature 1 = ', f01)

print(34603.0 - 21271.0)
print(21271 * 1.6147154870964945)
```

```python
# feature 2: number of trades change between 2 days in percent

#day 1
start_date = pd.to_datetime('2021-10-20 00:00:00')
end_date = pd.to_datetime('2021-10-20 23:00:00')
#day 2
start_date2 = pd.to_datetime('2021-10-21 00:00:00')
end_date2 = pd.to_datetime('2021-10-21 23:00:00')

#data

d01 = df.loc[(df['open_time'] > start_date) & (df['open_time'] < end_date)]
d02 = df.loc[(df['open_time'] > start_date2) & (df['open_time'] < end_date2)]

f02 = np.sum(d01['n_trades'])/np.sum(d02['n_trades'])*100

print('volumen Tag 1 = ', np.sum(d01['n_trades']))
print('volumen Tag 2 = ', np.sum(d02['n_trades']))
print('feature 2 = ', f02)

print(34603.0 - 21271.0)
print(21271 * 1.6147154870964945)
```

```python
# creating a dataframe 
d = {'f01 in %': [f01], 'f02 in %': [f02]}
df = pd.DataFrame(data=d)
df
```
